```python
def lyrics_crawling(name, path=''):
    driver = webdriver.Chrome('data_files/chromedriver.exe')
    driver.get('https://www.melon.com/index.htm')
    time.sleep(2)
    driver.find_element_by_css_selector('#top_search').send_keys(name)
    driver.find_element_by_css_selector('#gnb > fieldset > button.btn_icon.search_m').click()
    time.sleep(2)
    driver.find_element_by_css_selector('#frm_songList > div > table > tbody > tr > td:nth-child(3) > div > div > a.btn.btn_icon_detail').click()
    time.sleep(2)
    res = driver.page_source
    soup = BeautifulSoup(res, 'html.parser')
    text = soup.select('#d_video_summary')[0].text
    lyrics = ' '.join(re.compile('[ㄱ-ㅎ가-힣a-zA-Z0-9]+').findall(text))
    f = open(f'{path}{name}.txt', 'w')
    f.write(lyrics)
    f.close()
    driver.close()

    
    
lyrics_crawling('사랑은 늘 도망가 임영웅', path='lyrics/')
```
