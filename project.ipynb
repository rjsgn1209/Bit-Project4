```python
st_df = pd.read_csv('data_files/stopwords-ko.txt')
st_list = [st[0] for st in st_df.values]
st_list
```

```python
def remove_stwords_oneletter(doc):
  doc = okt.morphs(doc)
  doc = [t for t in doc if t not in st_list]
  doc = [t for t in doc if len(t) > 1]
  doc = ' '.join(doc)
  return doc
```

```python
df = pd.read_csv('data_files/전처리후 단어.csv', names=['단어', '슬픔', '분노', '기쁨', '긴장감', '무서움', '증오', '차분히', '섹시', '브릿락', '설렘', '몽환'])
```

```python
df.drop(0, inplace=True)
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].fillna(0).astype('int')
```

```python
df.set_index('단어', inplace=True)
```

```python
mood_dic = {1: '슬픔',
2: '분노',
3: '기쁨',
4: '긴장감',
5: '무서움',
6: '증오',
7: '차분히',
8: '섹시',
9: '브릿락',
10: '설렘',
11: '몽환'}
```

```python
df.loc['사치', '슬픔'] = 0
```

```python
df.head()
```

```python
from selenium import webdriver
driver = webdriver.Chrome('data_files/chromedriver.exe')
max_num = len(df.loc['차다':, :])
for i, word in enumerate(df.loc['차다':, :].index):
    
    print(f"{i}/{max_num} {word}", end='')
    url = f'https://ko.dict.naver.com/#/search?query={word}'
    driver.get(url)
    dic_key = input()
    if dic_key == '':
        df.drop(word, inplace=True)
        print(word, 'was deleted')
        print('=' * 50)
    else:
        dic_key = int(dic_key)
        print('set', mood_dic[dic_key])
        df.loc[word, mood_dic[dic_key]] = 1
        print('=' * 50)
```
