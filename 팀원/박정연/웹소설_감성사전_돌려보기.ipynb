```python
# 미친 중독


for i in range(1,2):
    url = f'https://novel.naver.com/webnovel/detail?novelId=994958&volumeNo={i}'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    lis = soup.select('div.detail_view_content')[0].text
    liss = re.compile('[ㄱ-ㅎ가-힣]+').findall(lis)
    print(liss)
    print('='*120)
    print('='*120)
    print('='*120)
```

# 토큰화 및 불용어 제거

```python
liss
```

```python
liss = [j for j in liss if len(j) >= 2]
```

```python
liss
```

```python
okt = Okt()

token=[]

for i in liss:
    token.append(okt.morphs(i))
    
    
```

```python
token
```

```python
tokenize = []

for tok in token:
    for to in tok:
        if len(to) >=2:
            tokenize.append(to)
        
```

# 웹소설 / 감성사전 적용시켜보기

```python
from collections import Counter
```

```python
for i in tokenize:
    for j in df.index:
        if i == j:
            print(i)
```

```python
# 단어별 개수

counter={}
for i in tokenize:
    for j in df.index:
        if i == j:
            if i in counter:
                counter[i] += 1
            else:
                counter[i] = 1
```

```python
counter
```

```python
df[df.index == '흔들리는']
```

```python
emotion_score = {}
emotion_word = {}
for key in counter.keys():
    for word in df.index:
        if key == word:
            for label in df.columns[df.loc[key,:] == 1]:
                if label in emotion_score:
                    emotion_score[label] += 1
                    emotion_word[label].append(word)
                else:
                    emotion_score[label] = 1
                    emotion_word[label] = [word]
                
                                    
            
```
